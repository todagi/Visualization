# Scatterplot(산점도)

#### 산점도는 연속형 이변량 자료를 시각화하는 기본 그래프
#### 두 변수 간 관계를 회귀에서 보고자 하는 경우에 수평축에 설명변수(x)를, 수직축에 반응변수(y)를 넣는다.
#### 일반적 상황에서는 가로 대 세로의 비인 aspect를 1로 설정해 그래프 영역이 정사각형이 되게 한다.

```r
es <- read.table("exam_score.txt", header = T)
head(es)
```

```r
str(es)
```

```r
attach(es)
plot(mid, final)
```

```r
summary(es)
```

#### final에 'NA'값이 있다. 이를 0점 처리한다.

```r
es$final[is.na(es$final)] <- 0
```

```r
sum(is.na(es$final))
```

```r
sum(is.na(es$mid))
```

```r
plot(es$mid, es$final)
```

```r
summary(es)
```

#### 제목과 점의 모양(pch : 1 ~ 69)
#### 축의 색깔 바꾸기 fg

```r
plot(mid, final, pch = 15, xlim = c(-5, 50), ylim = c(-5, 50), fg = "blue",
     col = "red", xlab = "mid exam", ylab = "final exam", main = "All exam score")
```

#### 자료점들이 격자에 찍혀 있는 경우가 있음. 이는 두 변수가 모두 정수값을 취하는 데서 기인함
#### 때문에 일부 점에 2개 이상의 개체가 쌓일 수 있는데 이는 흩뜨림(jittering)으로 해결함
#### 관측값에 임의오차를 붙임으로써 그래프에 n개의 점이 모두 찍히도록 함
#### x <= x + Ex, y <= y + Ey  (Ex, Ey는 -0.5와 0.5 사이의 균일분포 임의의 수)

```r
n <- length(es$mid)
plot(mid + runif(n, -0.5, 0.5), final + runif(n, -0.5, 0.5), pch = 19, xlim = c(-5, 50), ylim = c(-5, 50), fg = "green",
     col = rainbow(7), xlab = "mid exam", ylab = "final exam", main = "All exam score")
```

#### pch 점 모양

```r
x=rep(1:5,each=5);x
```

```r
y=rep(5:1,5);y
```

```r
plot(1:5,type="n",xlim=c(0,7.5),ylim=c(0,5.5))
points(x,y,pch=1:25,cex=2)
text(x-0.5,y,labels=as.character(1:25),cex=1.5)
points(rep(6,5),5:1,pch=65:69,cex=2)
text(rep(6,5)-0.5,5:1,labels=as.character(65:69),cex=1.5)
pchs=c("&","z","M","F","가")
points(rep(7,5),5:1,pch=pchs,cex=2,family="AppleGothic")
text(rep(7,5)-0.5,5:1,labels=pchs,cex=1.5,family="AppleGothic")
```
