## Stem-and-Leaf Plot(줄기잎그림)

#### 수치 자료의 분포를 보기 위하여 일차적으로 자료의 분포 특성을 살펴보기 위한 그림

```r
class <- c(81, 67, 92, 86, 94, 99, 54, 65, 83, 88, 
           91, 83, 78, 84, 94, 95, 68, 87, 85, 95,
           75, 80, 90, 84, 91, 76, 95, 77, 91, 65, 
           85, 68, 78, 94, 89, 85, 78, 87, 66, 72)
```

#### 81, 67, 92, 86
#### 10단위인 줄기(stem)와 1단위인 잎(leaf)로 분리

```r
#      줄기     잎
#       ↓       ↓
# 80 = 8 X 10 + 1
# 67 = 6 X 10 + 7
# 92 = 9 X 10 + 2
# 86 = 8 X 10 + 6
```

```r
6|7
7|
8|1 6
9|2
```

```r
stem(class)
```

```r
stem(class, scale = 2)   # scale 줄기 수. 줄기 수 2배 : scale = 2
```
