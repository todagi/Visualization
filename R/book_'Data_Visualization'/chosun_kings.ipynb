#### 줄 넣기

#### segments(x1, y1, x2, y2)  =>  점(x1, y1)과 점(x2, y2)를 잇는 선을 넣는다,  lwd 는 선 폭

```r
plot(c(0,0), c(28, 60), type = "n", xlim = c(0, 28), ylim = c(-5, 65),  
              xlab = "순서", ylab = "재위기간", main = "조선왕조")
points(1:27, chosun_kings$재위기간, pch = 15, col = "red")
segments(1:27, rep(0, 27), 1:27, chosun_kings$재위기간, lwd = 3, col = "red")
```

#### abline 수평선 : lty - 선 유형, lwd - 선 폭, col - 색깔

```r
plot(c(0,0), c(28, 60), type = "n", xlim = c(0, 28), ylim = c(-5, 65),  
              xlab = "순서", ylab = "재위기간", main = "조선왕조")
points(1:27, chosun_kings$재위기간, pch = 15, col = "red")
segments(1:27, rep(0, 27), 1:27, chosun_kings$재위기간, lwd = 3, col = "red")
abline(h = c(0, mean(chosun_kings$재위기간)), lty = "dotted", lwd = 1, col = "blue")
```

#### lines(x, y) 선 넣기 함수

```r
plot(c(0,0), c(28, 60), type = "n", xlim = c(0, 28), ylim = c(-5, 65),  
              xlab = "순서", ylab = "재위기간", main = "조선왕조")
points(1:27, chosun_kings$재위기간, pch = 15, col = "red")
segments(1:27, rep(0, 27), 1:27, chosun_kings$재위기간, lwd = 3, col = "red")
for(i in 1:27)
{lines(c(i,i), c(0, chosun_kings$재위기간[i]), lwd = 3, col = "blue")}
```

#### 사각형(rect), 다각형 넣기(polygon)

```r
windows(height = 5, width = 4.5)
plot(c(0, 25), c(0, 500), type = 'n', xlab = " ", ylab = " ", main = " ")
```

```r
plot(c(0, 25), c(0, 500), type = 'n', xlab = " ", ylab = " ", main = " ")
rect(5, 100, 10, 200, col = 'royalblue', border = 'royalblue')
rect(20, 400, 22.5, 450, col = 'royalblue', border = 'royalblue')
```

#### rect (5, 100, 10, 200) -> 왼쪽 아래꼭짓점 (5, 100)  오른쪽 위 꼭짓점 (10, 200)
#### rect (20, 400, 22.5, 450) -> 왼쪽 아래 꼭짓점이 (20, 400) 오른쪽 위 꼭짓점이 (22.5, 450)

#### rect(사각형) 응용

```r
period <- cumsum(chosun_kings$재위기간)
period
```

```r
plot(1:27, period, type = 'n', xlab = '순서', ylab = '누적연도', main = '조선왕조')
rect(0, 0, 1, period[1], col = 'royalblue', border = 'royalblue')
```

```r
plot(1:27, period, type = 'n', xlab = '순서', ylab = '누적연도', main = '조선왕조')
rect(0, 0, 1, period[1], col = 'royalblue', border = 'royalblue')
for(i in 2:27)
    {rect(i-1,period[i-1], i, period[i], col = 'royalblue', border = 'royalblue')}
segments(0, 0, 27, 518)
```

#### polygon 응용

#### polygon(c(0, 0, 1, 1), c(0, period[1], period[1], 0))
#### polygon(x, y) => 다각형의 m개의 꼭짓점의 x좌표와 y좌표 벡터

```r
period[1]
```

```r
plot(1:27, period, type = 'n', xlab = '순서', ylab = '누적연도', main = '조선왕조')
polygon(c(0,0,1,1), c(0,period[1], period[1], 0), col = rainbow(27)[1])
for(i in 2:27)
    {polygon(c(i-1,i-1,i,i), c(period[i-1], period[1], period[1], period[i-1]), col = rainbow(27)[i])}
segments(0, 0, 27, 518, lty = 'dotted')
```
