#### R 컬러

#### RGB(Red, Green, Blue) 코드로 표현. 각각 16진법의 두 자릿수(00부터 FF까지)로 컬러의 총수는 256^3(=16,777,216)가지 이다.
#### ex) red = #FF0000 / green = #00FF00 / blue = #0000FF / white(RGB가 완전히 채워짐) = #FFFFFF, black(RGB가 완전히 비어짐) = #000000

```r
colors <- c("#FF0000", "#FFFF00", "#00FF00", "#00FFFF", "#0000FF", "#FF00FF")
pie(rep(1, 6), col = colors, labels = colors)
```

```r
colors <- c("#FF0000", "#FFFF00", "#00FF00", "#00FFFF", "#0000FF", "#FF00FF")
pie(rep(1, 6), col = colors, labels = colors, main = 'red yellow green cyan blue magenta')
par(new = T)  # 겹쳐그리기
pie(rep(1, 1), col = "white", radius = 0.5, labels = " ")
```

#### grDevices 패키지(자동 설치) : rainbow, heat.colors, terrain.colors, topo.colors, cm.colors

```r
n <- 10
```

```r
barplot(rep(1, n), col = rainbow(n, alpha = 1), axes = F, main = "rainbow colors")
barplot(rep(1, n), col = heat.colors(n, alpha = 1), axes = F, main = "heat colors")
barplot(rep(1, n), col = terrain.colors(n, alpha = 1), axes = F, main = "terrain colors")
barplot(rep(1, n), col = topo.colors(n, alpha = 1), axes = F, main = "topo colors")
barplot(rep(1, n), col = cm.colors(n, alpha = 1), axes = F, main = "cyan-magenta colors")
```

#### alpha는 불투명도를 나타내며 0과 1사이의 값을 갖는다.

```r
rainbow(8, alpha = 1)
```

#### 순차형(sequentail) 컬러 계열

```r
display.brewer.all(type = "seq")
```

```r
barplot(rep(1, 9), col = brewer.pal(9, 'Reds'), axes = F, main = 'Brewer Reds')
barplot(rep(1, 9), col = brewer.pal(9, 'Greens'), axes = F, main = 'Brewer Greens')
```
